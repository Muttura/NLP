```python
import string


string.punctuation
```

```python
"I like NLP?"=="I like NLP"
```

```python
text="I like NLP.&$@"

def remove_punct(text):
    text1="".join([char for char in text if char not in string.punctuation])
    return text1
m=remove_punct(text)
print(m)
```

```python
df.columns
```

```python
df['body_text_clean']=df['Message'].apply(lambda x:remove_punct(x))
df.head()
```

```python
import re
```

```python
def tok(text):
    tokens=re.split('\W+',text)
    return tokens
```

```python
df['body_text_token']=df['body_text_clean'].apply(lambda x:tok(x.lower()))
df.head()
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stopword=nltk.corpus.stopwords.words('english')
print(stopword)
```

```python
def remove_stopword(tok_list):
    text=[word for word in tok_list if word not in stopword]
    return text
```

```python
df['body_text_nonstop']=df['body_text_token'].apply(lambda x:remove_stopword(x))
```
